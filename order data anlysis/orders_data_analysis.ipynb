```python
#renaming column name to lower case & also removing space from name
df.rename(columns = {'Order Id': 'order_id'})
#individual renaming of columns using rename()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
#comverted all column names to lower case & replaced the spacing to '_'
```

```python
#derive new columns discount, sale_price, 
df['discount'] = df['list_price']*df['discount_percent']*0.1
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
df.head(3)
```
